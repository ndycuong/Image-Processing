## **Low pass filter**
*Smoothing*

### **Mean filter**
*Replacing each pixel by the mean value of neighbor pixel*

```python
import cv2
from matplotlib import pyplot as plt
```

```python
pathName =r"C:\Users\nguye\OneDrive\code_anything\computerVision\test.jpg" #filenamef
image2 = cv2.imread(pathName, cv2.IMREAD_GRAYSCALE)
# The image will first be converted to grayscale
```

```python
figure_size = 9 #mean filter factor

new_image = cv2.blur(image2,(figure_size, figure_size))
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(new_image, cmap='gray'),plt.title('Mean filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

### **Gaussian filter**

```python
figure_size = 9 #Gaussian filter factor

new_image_gauss = cv2.GaussianBlur(image2, (figure_size, figure_size),0)
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(new_image_gauss, cmap='gray'),plt.title('Gaussian Filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

## **Laplacian filter**
*Sharpening filter*

```python
pathName =r"C:\Users\nguye\OneDrive\code_anything\computerVision\test2.jpg" #filenamef
image3 = cv2.imread(pathName, cv2.IMREAD_GRAYSCALE)
new_image1 = cv2.Laplacian(image3,cv2.CV_64F)

plt.figure(figsize=(11,6))
plt.subplot(131), plt.imshow(image3, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(new_image1, cmap='gray'),plt.title('Laplacian')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(image3 + new_image1, cmap='gray'),plt.title('adding original with laplacian image')
plt.xticks([]), plt.yticks([])
plt.show()
```

### Sharpening

```python
import numpy as np
from scipy.signal import convolve2d

# Sharpen
kernel = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

conv_im1 = convolve2d(image3,kernel)
plt.imshow((conv_im1), cmap='gray')
```

# **Non-linear filter**

## **Median filter**  
*Replacing each pixel by the median value of neighbor pixel*

```python
figure_size = 9 #Median filter factor

new_image = cv2.medianBlur(image2, figure_size)
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(new_image, cmap='gray'),plt.title('Median Filter')
plt.xticks([]), plt.yticks([])
plt.show()
```
