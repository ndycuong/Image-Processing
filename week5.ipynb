```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from scipy.signal import convolve2d

pathName =r"C:\Users\nguye\OneDrive\code_anything\computerVision\test8.jpg" #filenamef #test 5&8&9
image2 = cv2.imread(pathName, cv2.IMREAD_GRAYSCALE)

# Sharpen
Sobel = np.array([[0, 1, 2],
                    [-1, 0, 1],
                    [-2, -1, 0]])

Prewitt = np.array([[-1, -1, 0],
                    [-1, 0, 1],
                    [0, 1, 1]])

Laplace = np.array([[0, 1, 0],
                    [1, -8, 1],
                    [0, 1, 0]])


conv_im1 = convolve2d(image2,Sobel)
conv_im2 = convolve2d(image2,Prewitt)
conv_im3 = convolve2d(image2,Laplace)

fig, ax = plt.subplots(1,3,figsize=(30,30))
ax[0].imshow((conv_im1), cmap='gray')
ax[0].set_title('SOBEL')
ax[1].imshow((conv_im2), cmap='gray')
ax[1].set_title('PREWITT')
ax[2].imshow((conv_im3), cmap='gray')
ax[2].set_title('LAPLACE')
```
